```python
import pandas as pd
import polars as pl
from libs.io_lib.paths import DATA_DIR
```

```python
file_path = file_path = DATA_DIR /'datasets_symbols' /'symbol_0_merged.parquet'
```

```python
df = pd.read_parquet(file_path, engine='pyarrow')
```

```python
df
```

```python
df['responder_6']
```

```python
# Pour créer plusieurs lags (par exemple 5 lags)
for i in range(1, 150):
    df[f'responder_6_lag{i}'] = df['responder_6'].shift(i)

# Si vous voulez juste un lag spécifique, par exemple lag 1 :
df['responder_6_lag1'] = df['responder_6'].shift(1)
```

```python
df
```

```python
selected_columns = ['responder_6'] + [col for col in df.columns if col.startswith('responder_6_lag')]
df_selected = df[selected_columns]
```

```python
df_selected
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Création du plot
plt.figure(figsize=(10, 8))

# Créer le scatter plot avec une densité de points (kde)
sns.scatterplot(data=df_selected, 
                x='responder_6_lag1',  # Vous pouvez changer le lag ici (lag1, lag2, etc.)
                y='responder_6',
                alpha=0.1)  # Pour rendre les points plus transparents

# Calculer la corrélation
correlation = df_selected['responder_6'].corr(df_selected['responder_6_lag1'])

# Ajouter le titre et les labels
plt.title('Responder vs Its Lag')
plt.xlabel('lag 6')
plt.ylabel('responder 6')

# Ajouter la corrélation comme texte sur le plot
plt.text(0.7, 0.1, f'Correlation = {correlation:.2f}', 
         transform=plt.gca().transAxes, 
         fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Calculate correlation matrix
correlation_matrix = df_selected.corr()

# Create the plot
plt.figure(figsize=(12, 8))

# Create heatmap
sns.heatmap(correlation_matrix, 
            annot=True,  # Show numbers
            cmap='RdGy',  # Color scheme
            vmin=-1, vmax=1,  # Set limits for correlation values
            center=0,  # Center the colormap at 0
            fmt='.2f',  # Show 2 decimal places
            square=True)  # Make cells square

# Set title
plt.title('Correlation Matrix of Responder 6 and its Lags', pad=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show plot
plt.show()
```
